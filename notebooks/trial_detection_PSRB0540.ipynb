```python
import batanalysis as ba
import matplotlib.pyplot as plt
%matplotlib notebook
import os
os.environ['HEADASNOQUERY']='1' #This line is for dealing with heasoftpy outputs in jupyter notebooks

import numpy as np

#we will be querying from HEASARC so we will import this from astroquery
from astroquery.heasarc import Heasarc
heasarc=Heasarc()


##############.   Downloading the data at our preferred location...

object_name='PSR0540-69'
mission = 'swiftmastr'
table = heasarc.query_object(object_name, mission=mission, sortvar="START_TIME")
#table.pprint_all()
#print(table[:5])
obs_dir="/Users/slaha/Desktop/BAT_projects/GIT_Repository_directory_Jan2022/PSRB0540/"

ba.download_swiftdata(table[:10], save_dir=obs_dir) 
#Or esle it will save it in some place else: ~/.swift/swift_data_dir/



```

```python
################.  BAT Survey
obs_dir="/Users/slaha/Desktop/BAT_projects/GIT_Repository_directory_Jan2022/PSRB0540/"
batsurvey_obs=[]
input_dict=dict(obs_dir=obs_dir)       #Creating a dictionary to set up the input and output directories and the files
source_name="PSR0540-69"
source_id="PSR0540-69"

for i in table["OBSID"][:1]:

    print(i)
    try:
        
        obs=ba.BatSurvey(str(i),recalc=True,obs_dir=obs_dir)
    
        obs.save()
        batsurvey_obs.append(obs)
    except ValueError:
        print("Obsid has no survey data")



```

```python
source_name="PSR_B0540-69"
source_id="PSR_B0540-69"
for obs in batsurvey_obs:

    print("Running calculate_PHA for observation id", obs.pointing_ids)
    
    obs.merge_pointings()
    obs.calculate_pha(id_list=source_id, clean_dir=True)
    pha_list=obs.get_pha_filenames(id_list=source_name)
    ba.calc_response(pha_list)
    
    for pha in pha_list: 
        ba.fit_spectrum(pha, obs)
        print(pha)
        
            
```
